```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter

text = open('df1F.csv', 'r').read()

silhouette_image = Image.open('lolh2.png')

silhouette_image = silhouette_image.convert("L")  
silhouette_image = silhouette_image.filter(ImageFilter.GaussianBlur(2))  

silhouette_mask = np.array(silhouette_image)

wc = WordCloud(width=800, height=400, background_color='black', stopwords=STOPWORDS, mask=silhouette_mask, contour_width=1, contour_color='white').generate(text)

fig = plt.figure(figsize=(20, 10), dpi=300)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")

plt.savefig('wordcloud_high_res5.png', dpi=300, bbox_inches='tight', facecolor='black')

plt.show()


```
